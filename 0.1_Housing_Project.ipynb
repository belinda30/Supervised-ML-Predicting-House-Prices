### Imports

```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

from sklearn import set_config
#this code will make it so that whenever we transform data the output is a DataFrame
set_config(transform_output='pandas')
```

```python
sklearn.__version__
```

###Loading in the Data

```python
url = 'https://drive.google.com/file/d/1WsKRDZoNuX0AvE3qcOPYXZY_OQ_WbbAA/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]

data = pd.read_csv(path)
data.head(3)
```

###Define X and y

```python
X = data
y = X.pop('Expensive')
```

###Data Splitting (Train - Test)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3146)
```

### Impute missing data

```python
from sklearn.impute import SimpleImputer
```

Follow this approach for different types of estimators:

1. initialize,
2. use fit on the train set,
3. use transform on the train set,
4. use transform on the test set.
* (It is possible to combine steps 2 and 3 and use fit_transform instead.)

```python
# 1. Initialise
my_imputer = SimpleImputer()

# 2. use fit on the TRAIN set
my_imputer.fit(X_train)

# 3. use transform on the TRAIN set
X_imputed_train = my_imputer.transform(X_train)

# 4. use transform on the TEST set
X_imputed_test = my_imputer.transform(X_test)
```

### Modelling: Decision Tree

```python
# 1. import the model
from sklearn.tree import DecisionTreeClassifier

# 2. initialise the model (optional: give some parameters)
my_tree = DecisionTreeClassifier()

# 3. fit the model to the TRAIN data
my_tree.fit(X = X_imputed_train, y = y_train)
```

### Make Predictions and check accuracy on the TRAIN set

```python
from sklearn.metrics import accuracy_score
```

.predict() method: you pass in the input data (X_num_imputed_train) for which you want to predict the target variable (or class labels).

```python
# make predictions on TRAIN set
train_pred = my_tree.predict(X_imputed_train)
```

```python
train_pred
```

```python
# check accuracy on the TRAIN set with the predicted values

accuracy_score(y_true= y_train,
               y_pred= train_pred
               )
```

100%? That seems a LOT off

### Check accuracy for the TEST set

```python
# Make predictions on the TEST set

test_pred = my_tree.predict(X_imputed_test)
```

```python
# Check accuracy score on the TEST set

accuracy_score(y_true= y_test,
               y_pred= test_pred
              )
```

### Pipeline Creation

```python
from sklearn.pipeline import make_pipeline
```

```python
# Initialise transformer and model
imputer = SimpleImputer(strategy='median')
dtree = DecisionTreeClassifier(max_depth=4,
                               min_samples_leaf=10,
                               random_state=42)
```

```python
# Create a pipeline

pipe = make_pipeline(imputer, dtree)
```

```python
# fit the pipeline to the TRAIN data

pipe.fit(X_train, y_train)
```

```python
pipe.predict(X_test)
```

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 1. initialise transformers & model without specifying the parameters
imputer = SimpleImputer()
dtree= DecisionTreeClassifier()
```

```python
# 2. Create a pipeline
pipe = make_pipeline(imputer,
                     dtree)
```

```python
pipe
```

```python
# create parameter grid
param_grid = {
    'decisiontreeclassifier__max_depth': range(2,12),
    'decisiontreeclassifier__min_samples_leaf': range(3,10,2),
    'decisiontreeclassifier__min_samples_split': range(3,10,2),
    'simpleimputer__strategy': ['median', 'mean', 'most_frequent']
    }
```

```python
search = GridSearchCV(pipe,
                      param_grid,
                      cv=5,
                      scoring='accuracy',
                      verbose= 1)
```

```python
search.fit(X_train, y_train)
```

Explore best parameters and best score achieved with cross validation

```python
search.best_params_
```

```python
# the mean cross-validated score of the best estimator
search.best_score_
```

```python
# TRAIN set accuracy
train_pred = search.predict(X_train)

accuracy_score(y_train, train_pred)
```

```python
# TEST set accuracy
test_pred = search.predict(X_test)

accuracy_score(y_test, test_pred)
```

### Use GridSearchCv to find the best parameters of the pipeline

```python
from sklearn.preprocessing import StandardScaler
```

```python
# initialise transformers & model
imputer = SimpleImputer()
scaler = StandardScaler()
dtree= DecisionTreeClassifier()
```

```python
# create the pipeline
pipe = make_pipeline(imputer,
                     scaler,
                     dtree)
```

```python
pipe
```

```python
# create parameter grid
param_grid = {
    'simpleimputer__strategy': ['median', 'mean', 'most_frequent'],
    'standardscaler__with_mean': [True, False],
    'standardscaler__with_std': [True, False],
    'decisiontreeclassifier__max_depth': range(2,12),
    'decisiontreeclassifier__min_samples_leaf': range(3,10,2),
    'decisiontreeclassifier__min_samples_split': range(3,10,2)
    }
```

```python
# define cross validation
search = GridSearchCV(pipe,
                      param_grid,
                      cv=5,
                      verbose=1)
```

```python
# fit
search.fit(X_train, y_train)
```

```python
# cross validaion average accuracy
search.best_score_
```

```python
# best parameters
search.best_params_
```
